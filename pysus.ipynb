```python
# @title Importando o PySUS { display-mode: "form" }
from pysus.online_data import SINAN
import pandas as pd
```

```python
# @title Interagindo com o SINAN { run: "auto" }
import ipywidgets as widgets
from IPython.display import display
button = widgets.Button(description="Clique para mostrar lista de Agravos no SINAN!")
output = widgets.Output()
agrs = SINAN.list_diseases()

def on_button_clicked(b):
  # Display the message within the output widget.
  with output:
    for a in agrs:
      print(a)


button.on_click(on_button_clicked)
display(button, output)
```

```python
# @title Quais anos estão disponíveis? { run: "auto", display-mode: "form" }
Agravo = "deng" # @param ["deng", "mala"]
SINAN.get_available_years(Agravo)
```

```python
# @title Baixando os dados { run: "auto", display-mode: "form" }
Ano = 2011 # @param {type:"slider", min:2010, max:2022, step:1}
pth = SINAN.download(Agravo, Ano)
pd.read_parquet(pth)
```

## Descompactando um DBC
O PySUS já converte os DBC do DATASUS em arquivos parquet. Mas caso você tenha um arquivo DBC baixado diretamente do DATASUS, pode descompacta-lo com a função abaixo. depois é só salvar no formato que você quiser.

```python
from pysus.utilities.readdbc import read_dbc
```

```python
dados = read_dbc('dados.dbc')
dados.to_csv('dados.csv.gz')
```
